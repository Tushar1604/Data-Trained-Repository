# Column Description:

The countries here are ranked based, here a person rates his current life on a scale on which 10 is the best case and 0 being the worst.

    1. Country - Country name
    2. Region - Continent/ region where the country is located
    3. Happiness Rank - Rank of the Country based on the happiness score
    4. Happiness Score - Measure of the happiness the people
    5. Economy (GDP per Capita) - It is the measure of GDP per capita.
    6. Family - It is a measure of Support from Family, Relatives and Friends during difficult situations.
    7. Health (Life Expectancy) - Measure of health parameters and longetivity of life.
    8. Freedom - It is a measure of Freedom of choices.
    9. Trust (Government Corruption) - It is a measure of Trust in the government. It is a measure of degree of corruption in          the governement.
    10. Generosity - It is a measure of Generosity or Charity by an Individual of the Country.
    11. Dystopia Residual - It is the measure of unexplained parameters of the Happiness score.

In this Happiness Score is the Label - output after inference.

Country, Region are not inputs and are just for identification.

Happiness rank is the rank derived from Happiness Score. Hence it is neither a Feature nor a Label.

Features of the Data Set:

    1. Economy
    2. Family
    3. Health
    4. Freedom
    5. Trust
    6. Generosity

# Let's check the heat map to understand the Co-relation

A positive correlation indicates a strong dependency, while a negative correlation indicates a strong inverse dependency; 
a correlation coefficient closer to zero indicates weak dependence.



```python
plt.figure(figsize=(15,5))

sns.heatmap(df.corr(),annot = True, linewidths = 0.5, linecolor = "Black")
```

All the columns have positive corelation with the Happiness Score.

1. Max Corelation: GDP has the highest Corelation with Happiness Score (78%) followed by Family (74%) and Health (72%)

2. Min Corelation: Generosity has least corelation with Happiness Score (18%).

Health and Economy seems to have strong Co-relation with each other. This seems acceptable, since developed countries have better GDP per Capita and better health care facilities. Hence, we will consider it to be Data Related Multicollinearity and will not drop any Feature. We will continue with all the 6 Features.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
```

```python
lr = LinearRegression()
```

```python
x = df.drop(columns=['Happiness Score'])
x
```

```python
y = df['Happiness Score']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 355)
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
met = metrics.mean_squared_error(y_pred,y_test)
met
```
