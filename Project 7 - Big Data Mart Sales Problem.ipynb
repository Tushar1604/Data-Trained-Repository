# Inference

    There are a few columns with object data. We will need to Encode it.
    Label/ Target : Item_Outlet_Sales 
    

    1. Item_Visibility has a min value of zero. This is illogical as the item is being sold so there has to be visibility.
    2. Outlet_Establishment_Years vary from 1985 to 2009. Should be used as to how many years the store is active.
    3. The lower ‘count’ of Item_Weight, Outlet_size and Item_Outlet_Sales is relevant from the above null values

```python
df.apply(lambda x: len(x.unique()))
```

```python
#there are a total of 1559 products, and 16 different categories of products

# Now we can avoid checking the 'Item_Identifier, Outlet_identifier and source' columns
```

```python
#Finding frequency of categorical data while exckuding the Item_Identifier and Source data:

categorical_columns = [x for x in df.dtypes.index if df.dtypes[x]=='object']

#Exclude ID cols and source:
categorical_columns = [x for x in categorical_columns if x not in ['Item_Identifier','Outlet_Identifier','source']]

#Print frequency of categories
for col in categorical_columns:
    print ('\nFrequency of Categories for varible %s'%col)
    print (df[col].value_counts())
```

```python
# for fat content we can see that there is repitition of data, different names used for same category of data, therefore 
# combining the data:

#replacing low fat and LF with Low Fat
#replacing reg with Regular

df["Item_Fat_Content"].replace("low fat","Low Fat",inplace=True)
df["Item_Fat_Content"].replace("LF","Low Fat",inplace=True)
df["Item_Fat_Content"].replace("reg","Regular",inplace=True)
```

```python
df['Item_Fat_Content'].value_counts()
```

```python
#Average weight per item:
item_weight_avg=df.groupby(["Item_Identifier"]).agg({"Item_Weight":"mean"})
item_weight_avg
for index in item_weight_avg.index:
    null_condition=df.loc[df["Item_Identifier"]==index,["Item_Weight"]].isnull()
    for index_2 in null_condition.index:
        if null_condition.loc[index_2][0]==True:
            df.loc[index_2,["Item_Weight"]]=item_weight_avg.loc[index][0] 
```

```python
df["Outlet_Size"].value_counts()
```

```python
#Lets impute the Outlet_Size by mode of Outlet_Size groupped Outlet_Type

outlet_size_mode=df.groupby(["Outlet_Type"]).agg({"Outlet_Size":lambda x: x.value_counts().index[0]})
outlet_size_mode
for index in outlet_size_mode.index:
    null_condition=df.loc[df["Outlet_Type"]==index,["Outlet_Size"]].isnull()
    for index_2 in null_condition.index:
        if null_condition.loc[index_2][0]==True:
            df.loc[index_2,["Outlet_Size"]]=outlet_size_mode.loc[index][0]
```

```python
df.isnull().sum()
```

```python
df.sample(10)
```

```python
#Also one observation about the item visibility is that it can't be 0 so that also needs to be replaced


item_visibility_mean=df.groupby(["Item_Identifier"]).agg({"Item_Visibility":"mean"})
item_visibility_mean
for index in item_visibility_mean.index:
    zero_visibility_row=df.loc[(df["Item_Identifier"]==index)&(df["Item_Visibility"]==0)]
    if len(zero_visibility_row.index)>0:
        for index_2 in zero_visibility_row.index:
            df.loc[index_2,["Item_Visibility"]]=item_visibility_mean.loc[index][0]
```

```python
#Let us encode the data:

df1=pd.get_dummies(df,columns=["Item_Fat_Content","Item_Type","Outlet_Location_Type","Outlet_Size","Outlet_Type"])
```

```python
df1
```

```python
#lets check how old the store is:
df1["Outlet_Establishment_Year"]=2013-df1["Outlet_Establishment_Year"]
```

```python
df1["Outlet_Active Years"]=df1["Outlet_Establishment_Year"]
df1.drop(columns=["Outlet_Establishment_Year"],inplace=True)
```

```python
df1.sample(5)
```

```python
#lets chcek the skewness
skew_check=["Item_MRP","Item_Visibility","Item_Weight","Outlet_Active Years"]
for col in df1.skew().index:
    if col in skew_check:
        print("Skewness of column {} is: {}".format(col,df1.skew().loc[col]))
```

```python
#treating the skewness

df1.skew()
for col in df1.skew().index:
    if col in skew_check:
        if df1.skew().loc[col]>0.5:
            df1[col]=np.sqrt(df1[col])
        if df1.skew().loc[col]<-0.5:
            df1[col]=np.cbrt(df1[col])
```

```python
skew_check=["Item_MRP","Item_Visibility","Item_Weight","Outlet_Active Years"]
for col in df1.skew().index:
    if col in skew_check:
        print("Skewness of column {} is: {}".format(col,df1.skew().loc[col]))
```

```python
df1.drop(columns=["Item_Identifier","Outlet_Identifier"],inplace=True)
```

```python
df1.sample(5)
```

```python
#Divide into test and train:
df1_train = df1.loc[df1['source']=="df_train"]
df1_test = df1.loc[df1['source']=="df_test"]

#Drop unnecessary columns:
df1_test.drop(['Item_Outlet_Sales','source'],axis=1,inplace=True)
df1_train.drop(['source'],axis=1,inplace=True)


#Export files as modified versions:
df1_train.to_csv("train_modified.csv",index=False)
df1_test.to_csv("test_modified.csv",index=False)
```

```python
#seprating input and output from df_train

df1_x=df1_train.drop(columns=["Item_Outlet_Sales"])
y=df1_train[["Item_Outlet_Sales"]]
```

```python
#Lets bring every column to common scale

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x = sc.fit_transform(df1_x)
x = pd.DataFrame(x,columns=df1_x.columns)
```

```python
#to find random stat which gives maximum r2_score

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

def maxr2_score(regr,df1_x,y):
    max_r_score=0
    for r_state in range(35,60):
        x_train, x_test, y_train, y_test = train_test_split(df1_x, y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
#lets make a function which evaluates the model using cross_val_score

from sklearn.model_selection import cross_val_score
def model_evaluation(model,x,y):
    c_scores=cross_val_score(model,x,y,cv=5,scoring="r2")
    print("Mean r2 score for regressor: ",c_scores.mean())
    print("standard deviation in r2 score for regressor: ",c_scores.std())
    print(c_scores)
```

```python
#Lets use pca to reduce the dimensionality

from sklearn.decomposition import PCA
pca = PCA(n_components=23)
x_pca=pca.fit_transform(x)
print("Variance :{}".format(np.sum(pca.explained_variance_ratio_)))
```

```python

r_state=maxr2_score(dtr,x_pca,y)
```
